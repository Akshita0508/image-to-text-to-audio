```python
# --- Install dependencies (no Pillow downgrade = no restart needed) ---
!sudo apt-get install -y tesseract-ocr
!pip install pytesseract gTTS

# --- OCR + Text-to-Speech in Colab ---
from google.colab import files
from PIL import Image
import pytesseract
from gtts import gTTS
from IPython.display import Audio, display

# --- Step 1: Upload image ---
print("ðŸ“‚ Please upload an image (PNG/JPG)...")
uploaded = files.upload()

if uploaded:
    img_path = list(uploaded.keys())[0]
    print("âœ… Uploaded file:", img_path)

    # --- Step 2: OCR (Extract text) ---
    img = Image.open(img_path)
    text = pytesseract.image_to_string(img)

    print("\n--- Extracted Text ---\n")
    print(text if text.strip() else "âš ï¸ No text detected.")

    # --- Step 3: Convert Text â†’ Speech ---
    def text_to_speech(text, base_filename="output.mp3", lang="en"):
        max_len = 4500  # gTTS safe limit
        parts = [text[i:i+max_len] for i in range(0, len(text), max_len)]

        for idx, part in enumerate(parts):
            clean_part = " ".join(part.split())
            if clean_part.strip():
                fname = f"{base_filename.replace('.mp3','')}_{idx}.mp3"
                tts = gTTS(clean_part, lang=lang)
                tts.save(fname)
                print(f"âœ… Saved: {fname}")
                display(Audio(fname))

    if text.strip():
        text_to_speech(text)
    else:
        print("âŒ No readable text found in image.")
else:
    print("âš ï¸ No file uploaded. Please re-run this cell and upload again.")
```
